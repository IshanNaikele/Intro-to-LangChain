### Ollama Embeeding

```python
from langchain_community.embeddings import OllamaEmbeddings

embedding=(
    OllamaEmbeddings(model='mistral')
)


embedding
```

```python
r1=embedding.embed_documents(
    [
        "Ishan is a Clever Person",
        "I love my India",
        "Hey man,How are you ?",
        "Thank you ,Bro",
        "Ishan is a Clever Person"
    ]
)
```
